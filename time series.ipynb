```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
df = pd.read_csv('train.csv')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
train = df[df['Date']<='2020-05-10']
test = df.loc[(df['Date']>'2020-05-10')&(df['Date']<='2020-05-17')]
```

```python
Confirmed = train[train['Target']=='ConfirmedCases']
Confirmed = Confirmed.reset_index()
Confirmed.head()
```

```python
Country_Region = list(Confirmed['Country_Region'].unique())
#print(Country_Region)
```

```python
def first_non_zero(mylist):
    for index, number in enumerate(mylist):
        if number >0: 
            return index
```

```python
Afghanistan = Confirmed[Confirmed['Country_Region']=='Afghanistan']
Afghanistan = Afghanistan.iloc[first_non_zero(Afghanistan['TargetValue']):]
Afghanistan.Timestamp = pd.to_datetime(Afghanistan['Date'],format='%Y-%m-%d')
Afghanistan.index = Afghanistan.Timestamp
Afghanistan
```

```python
test_valid = test.loc[(test['Country_Region']=='Afghanistan')& (test['Target']=='ConfirmedCases')&(test['Date']<='2020-05-17')]
test_valid.Timestamp = pd.to_datetime(test_valid['Date'],format='%Y-%m-%d')
test_valid.index = test_valid.Timestamp
```

```python
from statsmodels.tsa.api import ExponentialSmoothing
import matplotlib.pyplot as plt

y_hat_avg = test_valid.copy()

fit1 = ExponentialSmoothing(Afghanistan['TargetValue'] ,seasonal_periods=7 ,trend='add', seasonal='add',damped=True).fit()
y_hat_avg['Holt_Winter'] = fit1.forecast(len(test_valid))
#plt.figure(figsize=(16,8))
#plt.plot(Afghanistan['TargetValue'], label='Train')
#plt.plot(test_valid['TargetValue'], label='test')
#plt.plot(y_hat_avg['Holt_Winter'], label='predict')
#plt.legend(loc='best')
#plt.show()
```

```python
print(y_hat_avg['Holt_Winter'])
```

```python
columns = ['Country_Region','Predict Value']
date = pd.date_range('2020-05-11',periods = 7)
df_result = pd.DataFrame(index = date,columns = columns)
result = []
Country_Region = list(Confirmed['Country_Region'].unique())
for i in Country_Region:
    datai = Confirmed[Confirmed['Country_Region']==i]
    datai = datai.iloc[first_non_zero(datai['TargetValue']):]
    datai.Timestamp = pd.to_datetime(datai['Date'],format='%Y-%m-%d')
    datai.index = datai.Timestamp
    fiti = ExponentialSmoothing(datai['TargetValue'] ,seasonal_periods=10,trend='add', seasonal='add',damped=True).fit()
    #df_result['Country_Region'] = i
    #df_result['Predict Value'] = fiti.forecast(7)
    result.append(list(fiti.forecast(7)))
#df_result
```

```python
from pandas.core.frame import DataFrame
c={"Predict Value" : result}
result_df=DataFrame(c)
result_df['Country_Region'] = Country_Region
result_df
```

```python
Fatalities = train[train['Target']=='Fatalities']
Fatalities = Fatalities.reset_index()
Fatalities.head()
```

```python
Afghanistan_f = Fatalities[Fatalities['Country_Region']=='Afghanistan']
Afghanistan_f = Afghanistan_f.iloc[first_non_zero(Afghanistan_f['TargetValue']):]
Afghanistan_f.Timestamp = pd.to_datetime(Afghanistan_f['Date'],format='%Y-%m-%d')
Afghanistan_f.index = Afghanistan_f.Timestamp
```

```python
test_f = test.loc[(test['Country_Region']=='Afghanistan')& (test['Target']=='Fatalities')]
test_f.Timestamp = pd.to_datetime(test_f['Date'],format='%Y-%m-%d')
test_f.index = test_f.Timestamp
```

```python
from statsmodels.tsa.api import ExponentialSmoothing
import matplotlib.pyplot as plt

f_hat_avg = test_f.copy()

fit2 = ExponentialSmoothing(Afghanistan_f['TargetValue'] ,seasonal_periods=7 ,trend='add', seasonal='add',damped=True).fit()
f_hat_avg['Holt_Winter'] = fit2.forecast(len(test_f))
print(f_hat_avg['Holt_Winter'])
```

```python
f_result = []
Country_Region = list(Fatalities['Country_Region'].unique())
for i in Country_Region:
    datai = Fatalities[Fatalities['Country_Region']==i]
    datai = datai.iloc[first_non_zero(datai['TargetValue']):]
    datai.Timestamp = pd.to_datetime(datai['Date'],format='%Y-%m-%d')
    datai.index = datai.Timestamp
    fiti = ExponentialSmoothing(datai['TargetValue'] ,seasonal_periods=4,trend='add', seasonal='add',damped=True).fit()
    #df_result['Country_Region'] = i
    #df_result['Predict Value'] = fiti.forecast(7)
    f_result.append(list(fiti.forecast(7)))
#df_result
```

```python
f={"Predict Value" : f_result}
f_result_df=DataFrame(f)
f_result_df['Country_Region'] = Country_Region
f_result_df
```
